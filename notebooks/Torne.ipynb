```python
from tqdm import tqdm
from diffusers.optimization import get_cosine_schedule_with_warmup
# Define reconstruction functions with loss tracking
import copy

def run_mumott_reconstruction(dc, ell_max, maxiter, reg_weight, verbose=True):
    """Run mumott LBFGS reconstruction and return results with timing."""
    basis_set = SphericalHarmonics(ell_max=ell_max)
    projector = SAXSProjector(dc.geometry)  # Use CPU version
    
    residual_calculator = GradientResidualCalculator(
        data_container=dc,
        basis_set=basis_set,
        projector=projector
    )
    
    loss_function = SquaredLoss(residual_calculator)
    if reg_weight > 0:
        regularizer = Laplacian()
        loss_function.add_regularizer(
            name='laplacian',
            regularizer=regularizer,
            regularization_weight=reg_weight
        )
    
    optimizer = LBFGS(loss_function, maxiter=maxiter)
    
    if verbose:
        print(f"Running mumott LBFGS ({maxiter} iterations)...")
    
    start = time.time()
    results = optimizer.optimize()
    total_time = time.time() - start
    
    # Get reconstruction
    reconstruction = residual_calculator.coefficients.copy()
    
    # Get final loss - get_loss() returns a dict, extract the loss term
    final_loss_dict = loss_function.get_loss()
    if isinstance(final_loss_dict, dict):
        # Get the squared_loss or first non-None value
        final_loss = final_loss_dict.get('squared_loss', None)
        if final_loss is None:
            for v in final_loss_dict.values():
                if v is not None:
                    final_loss = v
                    break
        if final_loss is None:
            final_loss = 0.0
    else:
        final_loss = final_loss_dict
    
    return {
        'reconstruction': reconstruction,
        'time': total_time,
        'final_loss': final_loss,
        'iterations': maxiter
    }


def run_pytorch_reconstruction(dc, ell_max, maxiter, lr, reg_weight, reg_type='laplacian', volume=None, verbose=True, batch_size=None):
    """Run PyTorch AdamW reconstruction and return results with timing.
    
    Args:
        dc: DataContainer with projections and geometry
        ell_max: Maximum spherical harmonic order
        maxiter: Number of optimization iterations
        lr: Learning rate
        reg_weight: Regularization weight
        reg_type: Type of regularization ('laplacian' or 'none')
        volume: Optional initial volume tensor
        verbose: Whether to print progress
        batch_size: Number of projections to use per iteration. If None, use all projections.
                   If specified, a random subset of projections is selected each iteration.
    """
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    geometry = dc.geometry
    volume_shape = geometry.volume_shape
    num_coeffs = (ell_max // 2 + 1) * (ell_max + 1)
    total_projections = len(dc.projections)
    
    # Get full target data
    full_target_data = torch.tensor(
        np.stack([p.data for p in dc.projections]).astype(np.float32),
        device=device, dtype=torch.float32
    )
    
    # Create basis set and projection matrix
    basis_set = SphericalHarmonics(ell_max=ell_max, probed_coordinates=geometry.probed_coordinates)
    projection_matrix = torch.tensor(
        basis_set.projection_matrix.astype(np.float32),
        device=device, dtype=torch.float32
    )
    
    # Initialize volume
    if volume is None:
        volume = torch.zeros(
            (*volume_shape, num_coeffs),
            device=device, dtype=torch.float32, requires_grad=True
        )
    else:
        volume = volume.to(device).clone().detach().requires_grad_(True)
    
    # Optimizer
    optimizer = torch.optim.AdamW([volume], lr=lr, weight_decay=0.0)
    # scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    #     optimizer, mode='min', factor=0.1, patience=20,
    #     threshold=1e14,  # Or threshold_mode='rel', threshold=0.01 for 1% relative change
    #     threshold_mode='abs'  # 'abs' for absolute change, 'rel' for relative
    #     )
    # scheduler = torch.optim.lr_scheduler.StepLR(
    #     optimizer, step_size=100, gamma=0.2
    # )
    # cosine scheduler with warm up
    scheduler = get_cosine_schedule_with_warmup(
        optimizer=optimizer,
        num_warmup_steps=50,
        num_training_steps=maxiter,
    )
    
    losses = []
    
    # Determine if we're using mini-batches
    use_minibatch = batch_size is not None and batch_size < total_projections
    
    if verbose:
        batch_info = f", batch_size={batch_size}" if use_minibatch else " (full batch)"
        print(f"Running PyTorch AdamW ({maxiter} iterations{batch_info})...")
        if use_minibatch:
            print(f"  Total projections: {total_projections}, using {batch_size} per iteration")
    
    torch.cuda.synchronize() if torch.cuda.is_available() else None
    start = time.time()
    
    iterator = tqdm(range(maxiter), disable=not verbose)
    for i in iterator:
        optimizer.zero_grad()
        
        # Select projections for this iteration
        if use_minibatch:
            # Randomly select batch_size projection indices
            batch_indices = np.random.choice(total_projections, size=batch_size, replace=False)
            batch_indices = np.sort(batch_indices)  # Sort for consistent ordering
            
            # Create a deep copy of dc and modify it for this batch
            batch_dc = copy.deepcopy(dc)
            
            # Delete projections not in batch (in reverse order to preserve indices)
            indices_to_delete = [idx for idx in range(total_projections) if idx not in batch_indices]
            for idx in sorted(indices_to_delete, reverse=True):
                del batch_dc.projections[idx]
            
            # Build projector for this batch using the modified dc's geometry
            batch_projector = build_mumott_projector(batch_dc.geometry, device=device)
            
            # Get corresponding target data
            target_data = full_target_data[batch_indices]
            current_geometry = batch_dc.geometry
        else:
            # Use full batch - build projector once on first iteration
            if i == 0:
                full_projector = build_mumott_projector(geometry, device=device)
            batch_projector = full_projector
            target_data = full_target_data
            current_geometry = geometry
        
        # Forward pass
        spatial_proj = batch_projector(volume)
        
        predicted = forward_quadrature(
            current_geometry.probed_coordinates,
            spatial_proj,
            ell_max=ell_max,
            # projection_matrix=projection_matrix,
            # mumott_compat=True,  # Apply scaling factor to match mumott's output
            # mode='simpson'
        )
        
        # Compute loss
        data_loss = torch.nn.functional.mse_loss(predicted, target_data, reduction='sum')
        
        # Regularization
        reg_loss = torch.tensor(0.0, device=device)
        if reg_weight > 0 and reg_type == 'laplacian':
            # 3D Laplacian regularization
            laplacian = (
                torch.roll(volume, 1, 0) + torch.roll(volume, -1, 0) +
                torch.roll(volume, 1, 1) + torch.roll(volume, -1, 1) +
                torch.roll(volume, 1, 2) + torch.roll(volume, -1, 2) - 6 * volume
            )
            reg_loss = reg_weight * torch.sum(laplacian ** 2)
        
        loss = data_loss + reg_loss
        losses.append(loss.item())
        
        # Backward pass
        loss.backward()
        
        # Gradient clipping
        torch.nn.utils.clip_grad_norm_([volume], max_norm=1.0)
        
        optimizer.step()
        scheduler.step(loss)
        
        iterator.set_postfix({"status":f"Iter {i+1:3d}: loss={loss.item():.3e}"})
    
    torch.cuda.synchronize() if torch.cuda.is_available() else None
    total_time = time.time() - start
    
    return {
        'reconstruction': volume.detach().cpu(),
        'time': total_time,
        'final_loss': losses[-1],
        'losses': losses,
        'iterations': maxiter
    }

print("Reconstruction functions defined.")
```

```python
def run_pytorch_reconstruction_native(dc, ell_max, maxiter, lr, reg_weight, reg_type='laplacian', 
                                       volume=None, verbose=True, batch_size=None, proj_batch_size=8,
                                       warmup_steps=0):
    """Run PyTorch AdamW reconstruction using the native PyTorchProjectorOptimized.
    
    This version uses the native PyTorch projector which is ~11x faster than ASTRA
    and fully differentiable via torch.autograd.
    
    Args:
        dc: DataContainer with projections and geometry
        ell_max: Maximum spherical harmonic order
        maxiter: Number of optimization iterations
        lr: Learning rate (target learning rate after warmup)
        reg_weight: Regularization weight
        reg_type: Type of regularization ('laplacian', 'sh', or 'none')
                  - 'laplacian': 3D Laplacian smoothness
                  - 'sh': Spherical Harmonics (penalizes high angular frequencies by ℓ(ℓ+1))
                  - 'none': No regularization
        volume: Optional initial volume tensor
        verbose: Whether to print progress
        batch_size: Number of projections to use per iteration. If None, use all projections.
                   If specified, a random subset of projections is selected each iteration.
        proj_batch_size: Batch size for the PyTorch projector (default: 8)
        warmup_steps: Number of iterations for linear learning rate warmup (default: 0)
                     LR increases linearly from lr/warmup_steps to lr over warmup_steps iterations
    
    Returns:
        dict with reconstruction, time, final_loss, losses, iterations
    """
    from smartt.projectors.pytorch_projector import PyTorchProjectorOptimized
    from smartt.shutils.evaulate_sh import forward_quadrature
    
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    geometry = dc.geometry
    volume_shape = geometry.volume_shape
    num_coeffs = (ell_max // 2 + 1) * (ell_max + 1)
    total_projections = len(dc.projections)
    
    # Get full target data
    full_target_data = torch.tensor(
        np.stack([p.data for p in dc.projections]).astype(np.float32),
        device=device, dtype=torch.float32
    )
    
    # Create basis set for forward_quadrature
    basis_set = SphericalHarmonics(ell_max=ell_max, probed_coordinates=geometry.probed_coordinates)
    
    # Initialize volume
    if volume is None:
        volume = torch.zeros(
            (*volume_shape, num_coeffs),
            device=device, dtype=torch.float32, requires_grad=True
        )
    else:
        volume = volume.to(device).clone().detach().requires_grad_(True)
    
    # Optimizer (start with lr/max(1, warmup_steps) if using warmup)
    initial_lr = lr / max(1, warmup_steps) if warmup_steps > 0 else lr
    optimizer = torch.optim.AdamW([volume], lr=initial_lr, weight_decay=0.0)
    # scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
    #     optimizer, mode='min', factor=0.1, patience=20,
    #     threshold=1e13, threshold_mode='abs'
    # )

    scheduler = get_cosine_schedule_with_warmup(
        optimizer=optimizer,
        num_warmup_steps=50,
        num_training_steps=maxiter,
    )
    
    losses = []
    
    # Determine if we're using mini-batches (projection selection)
    use_minibatch = batch_size is not None and batch_size < total_projections
    
    if verbose:
        batch_info = f", batch_size={batch_size}" if use_minibatch else " (full batch)"
        print(f"Running PyTorch Native Reconstruction ({maxiter} iterations{batch_info})...")
        print(f"  Using PyTorchProjectorOptimized (proj_batch_size={proj_batch_size})")
        print(f"  Regularization: {reg_type} (weight={reg_weight})")
        if warmup_steps > 0:
            print(f"  Warmup: {warmup_steps} steps (lr: {initial_lr:.2e} -> {lr:.2e})")
        else:
            print(f"  Learning rate: {lr:.2e}")
        if use_minibatch:
            print(f"  Total projections: {total_projections}, using {batch_size} per iteration")
    
    # Build projector once for full geometry (we'll handle mini-batches differently if needed)
    if not use_minibatch:
        projector = PyTorchProjectorOptimized(geometry, device=device)
    
    torch.cuda.synchronize() if torch.cuda.is_available() else None
    start = time.time()
    
    iterator = tqdm(range(maxiter), disable=not verbose)
    for i in iterator:
        # Linear warmup: increase LR from initial_lr to lr over warmup_steps
        if warmup_steps > 0 and i < warmup_steps:
            warmup_factor = (i + 1) / warmup_steps
            current_lr = initial_lr + warmup_factor * (lr - initial_lr)
            for param_group in optimizer.param_groups:
                param_group['lr'] = current_lr
        
        optimizer.zero_grad()
        
        # Select projections for this iteration
        if use_minibatch:
            # Randomly select batch_size projection indices
            batch_indices = np.random.choice(total_projections, size=batch_size, replace=False)
            batch_indices = np.sort(batch_indices)
            
            # Create a deep copy of dc and modify it for this batch
            batch_dc = copy.deepcopy(dc)
            indices_to_delete = [idx for idx in range(total_projections) if idx not in batch_indices]
            for idx in sorted(indices_to_delete, reverse=True):
                del batch_dc.projections[idx]
            
            # Build projector for this batch
            batch_projector = PyTorchProjectorOptimized(batch_dc.geometry, device=device)
            
            # Get corresponding target data
            target_data = full_target_data[batch_indices]
            current_geometry = batch_dc.geometry
        else:
            batch_projector = projector
            target_data = full_target_data
            current_geometry = geometry
        
        # Forward pass: spatial projection using native PyTorch projector
        spatial_proj = batch_projector(volume, batch_size=proj_batch_size)
        
        # Forward quadrature to get final prediction
        predicted = forward_quadrature(
            current_geometry.probed_coordinates,
            spatial_proj,
            ell_max=ell_max,
        )
        
        # Compute loss
        data_loss = torch.nn.functional.mse_loss(predicted, target_data, reduction='sum')
        
        # Regularization
        reg_loss = torch.tensor(0.0, device=device)
        if reg_weight > 0:
            if reg_type == 'laplacian':
                # 3D Laplacian regularization
                laplacian = (
                    torch.roll(volume, 1, 0) + torch.roll(volume, -1, 0) +
                    torch.roll(volume, 1, 1) + torch.roll(volume, -1, 1) +
                    torch.roll(volume, 1, 2) + torch.roll(volume, -1, 2) - 6 * volume
                )
                reg_loss = reg_weight * torch.sum(laplacian ** 2)
            elif reg_type == 'sh':
                # Spherical Harmonics regularization
                # Penalize coefficients by ℓ(ℓ+1) where ℓ is the SH degree
                # This is the natural regularization because SH are eigenfunctions of 
                # the Laplace-Beltrami operator: ∇²_S Y_ℓ^m = -ℓ(ℓ+1) Y_ℓ^m
                
                from smartt.shutils.evaulate_sh import _generate_lm_list
                lm_list = _generate_lm_list(ell_max)
                
                # Build weight vector: ℓ(ℓ+1) for each coefficient
                weights = torch.tensor(
                    [l * (l + 1) for l, m in lm_list],
                    device=device, dtype=torch.float32
                )  # Shape: (num_coeffs,)
                
                # Apply weights to each coefficient channel
                # volume shape: (X, Y, Z, num_coeffs)
                weighted_coeffs = volume ** 2 * weights[None, None, None, :]
                sh_loss = torch.sum(weighted_coeffs)
                
                reg_loss = reg_weight * sh_loss
        
        loss = data_loss + reg_loss
        losses.append(loss.item())
        
        # Backward pass
        loss.backward()
        
        # Gradient clipping
        torch.nn.utils.clip_grad_norm_([volume], max_norm=1.0)
        
        optimizer.step()
        scheduler.step(loss)
        
        # Update progress bar with loss and current LR
        current_lr = optimizer.param_groups[0]['lr']
        iterator.set_postfix({"loss": f"{loss.item():.3e}", "lr": f"{current_lr:.2e}"})
    
    torch.cuda.synchronize() if torch.cuda.is_available() else None
    total_time = time.time() - start
    
    return {
        'reconstruction': volume.detach().cpu(),
        'time': total_time,
        'final_loss': losses[-1],
        'losses': losses,
        'iterations': maxiter
    }

print("Native PyTorch reconstruction function defined (with SH regularization + warmup support).")
```

```python
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

def sh_l2_to_tensor(sh_coeffs):
    """
    Convert ℓ=2 spherical harmonic coefficients to a 3x3 symmetric traceless tensor.
    
    The ℓ=2 SH coefficients (5 components: m = -2, -1, 0, 1, 2) correspond to indices 1-5
    in the full SH array (assuming ℓ=0 is index 0).
    
    The conversion uses the standard real spherical harmonics convention:
    Y_2^{-2}, Y_2^{-1}, Y_2^0, Y_2^1, Y_2^2 → tensor components
    
    Args:
        sh_coeffs: array of shape (..., n_sh) where n_sh >= 6 (includes ℓ=0 and ℓ=2)
    
    Returns:
        tensor: array of shape (..., 3, 3) symmetric traceless tensor
    """
    # Extract ℓ=2 coefficients (indices 1-5, assuming ordering: ℓ=0 at 0, ℓ=2 at 1-5)
    # Standard real SH ordering: Y_2^{-2}, Y_2^{-1}, Y_2^0, Y_2^1, Y_2^2
    c = sh_coeffs[..., 1:6]  # 5 components for ℓ=2
    
    # Normalization factor (depends on convention)
    sqrt3 = np.sqrt(3.0)
    sqrt15 = np.sqrt(15.0)
    sqrt5 = np.sqrt(5.0)
    
    # Build the symmetric traceless tensor
    # Using the standard conversion from real ℓ=2 SH to Cartesian tensor:
    # T_xx, T_yy, T_zz, T_xy, T_xz, T_yz
    
    shape = sh_coeffs.shape[:-1]
    tensor = np.zeros(shape + (3, 3), dtype=sh_coeffs.dtype)
    
    # Y_2^0 ~ (3z^2 - r^2) / (2*sqrt(5)) → contributes to diagonal
    # Y_2^{-2} ~ xy, Y_2^{-1} ~ yz, Y_2^1 ~ xz, Y_2^2 ~ (x^2 - y^2)
    
    # Standard mapping (real SH to tensor components):
    # c[0] = Y_2^{-2} → T_xy
    # c[1] = Y_2^{-1} → T_yz  
    # c[2] = Y_2^0   → T_zz (and traceless diagonal)
    # c[3] = Y_2^1   → T_xz
    # c[4] = Y_2^2   → T_xx - T_yy
    
    # T_xy component
    tensor[..., 0, 1] = c[..., 0] * np.sqrt(15.0 / 4.0)
    tensor[..., 1, 0] = tensor[..., 0, 1]
    
    # T_yz component
    tensor[..., 1, 2] = c[..., 1] * np.sqrt(15.0 / 4.0)
    tensor[..., 2, 1] = tensor[..., 1, 2]
    
    # T_xz component  
    tensor[..., 0, 2] = c[..., 3] * np.sqrt(15.0 / 4.0)
    tensor[..., 2, 0] = tensor[..., 0, 2]
    
    # Diagonal: T_zz from Y_2^0, and (T_xx - T_yy) from Y_2^2
    # Y_2^0 = sqrt(5/16π) * (3z^2 - r^2) → traceless: T_zz = 2*c0, T_xx = T_yy = -c0
    # Y_2^2 = sqrt(15/16π) * (x^2 - y^2) → T_xx - T_yy = 2*c4
    
    c0_contrib = c[..., 2] * np.sqrt(5.0 / 4.0)
    c2_contrib = c[..., 4] * np.sqrt(15.0 / 4.0) / 2.0
    
    tensor[..., 2, 2] = 2.0 * c0_contrib / 3.0      # T_zz
    tensor[..., 0, 0] = -c0_contrib / 3.0 + c2_contrib  # T_xx
    tensor[..., 1, 1] = -c0_contrib / 3.0 - c2_contrib  # T_yy
    
    return tensor


def get_principal_direction(sh_volume):
    """
    Compute the principal (dominant) direction for each voxel from ℓ=2 SH coefficients.
    
    Args:
        sh_volume: array of shape (nx, ny, nz, n_sh) with spherical harmonic coefficients
    
    Returns:
        directions: array of shape (nx, ny, nz, 3) with principal direction (unit vector)
        eigenvalues: array of shape (nx, ny, nz, 3) with eigenvalues (sorted descending)
    """
    # Convert to numpy if tensor
    if hasattr(sh_volume, 'cpu'):
        sh_volume = sh_volume.cpu().detach().numpy()
    
    # Get tensor representation
    tensors = sh_l2_to_tensor(sh_volume)
    
    # Compute eigenvalues and eigenvectors for each voxel
    # eigh returns eigenvalues in ascending order
    eigenvalues, eigenvectors = np.linalg.eigh(tensors)
    
    # Principal eigenvector is the one with largest eigenvalue (last column)
    principal_directions = eigenvectors[..., -1]  # shape: (nx, ny, nz, 3)
    
    # Sort eigenvalues descending
    eigenvalues_sorted = eigenvalues[..., ::-1]
    
    return principal_directions, eigenvalues_sorted, eigenvalues


def direction_to_rgb(directions):
    """
    Convert direction vectors to RGB colors using absolute value mapping.
    Maps |x|, |y|, |z| → R, G, B (common for fiber orientation visualization).
    
    Args:
        directions: array of shape (..., 3) with direction vectors
    
    Returns:
        rgb: array of shape (..., 3) with RGB values in [0, 1]
    """
    # Take absolute value (directions are symmetric: v and -v are equivalent)
    abs_dirs = np.abs(directions)
    # Normalize to unit length (in case they aren't already)
    norms = np.linalg.norm(abs_dirs, axis=-1, keepdims=True)
    norms = np.where(norms > 1e-10, norms, 1.0)  # Avoid division by zero
    rgb = abs_dirs / norms
    return rgb


@interact(proj_num=(0, native_results['reconstruction'].shape[0] - 1))
def show_dominant_direction(proj_num):
    """Interactive visualization of dominant direction from ℓ=2 spherical harmonics."""
    
    # Get reconstructions
    native_vol = native_results['reconstruction']
    pytorch_vol = pytorch_results['reconstruction']
    
    # Compute principal directions for each slice
    native_dirs, native_eigs, _ = get_principal_direction(native_vol[proj_num])
    pytorch_dirs, pytorch_eigs, _ = get_principal_direction(pytorch_vol[proj_num])
    
    # Convert to RGB for visualization
    native_rgb = direction_to_rgb(native_dirs)
    pytorch_rgb = direction_to_rgb(pytorch_dirs)
    
    # Create figure
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    
    # Row 1: Native results
    axes[0, 0].imshow(native_rgb)
    axes[0, 0].set_title(f'Native - Direction RGB (slice {proj_num})')
    axes[0, 0].axis('off')
    
    im1 = axes[0, 1].imshow(native_eigs[..., 0], cmap='viridis')
    axes[0, 1].set_title('Native - Largest Eigenvalue')
    axes[0, 1].axis('off')
    plt.colorbar(im1, ax=axes[0, 1], fraction=0.046)
    
    # Anisotropy measure: (λ1 - λ3) / (λ1 + λ2 + λ3 + eps)
    native_aniso = (native_eigs[..., 0] - native_eigs[..., 2]) / (np.sum(native_eigs, axis=-1) + 1e-10)
    im2 = axes[0, 2].imshow(native_aniso, cmap='hot', vmin=0)
    axes[0, 2].set_title('Native - Anisotropy')
    axes[0, 2].axis('off')
    plt.colorbar(im2, ax=axes[0, 2], fraction=0.046)
    
    # Row 2: PyTorch results
    axes[1, 0].imshow(pytorch_rgb)
    axes[1, 0].set_title(f'PyTorch - Direction RGB (slice {proj_num})')
    axes[1, 0].axis('off')
    
    im3 = axes[1, 1].imshow(pytorch_eigs[..., 0], cmap='viridis')
    axes[1, 1].set_title('PyTorch - Largest Eigenvalue')
    axes[1, 1].axis('off')
    plt.colorbar(im3, ax=axes[1, 1], fraction=0.046)
    
    pytorch_aniso = (pytorch_eigs[..., 0] - pytorch_eigs[..., 2]) / (np.sum(pytorch_eigs, axis=-1) + 1e-10)
    im4 = axes[1, 2].imshow(pytorch_aniso, cmap='hot', vmin=0)
    axes[1, 2].set_title('PyTorch - Anisotropy')
    axes[1, 2].axis('off')
    plt.colorbar(im4, ax=axes[1, 2], fraction=0.046)
    
    plt.suptitle('Dominant Direction from ℓ=2 Spherical Harmonics\n(RGB: |x|→R, |y|→G, |z|→B)', fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
# ============================================================================
# Compare forward model consistency
# ============================================================================
print("\n" + "=" * 80)
print("3. Forward Model Consistency Check")
print("=" * 80)

def compute_metrics(pred, target, name):
    """Compute various metrics between prediction and target."""
    diff = pred - target
    mse = np.mean(diff ** 2)
    rmse = np.sqrt(mse)
    mae = np.mean(np.abs(diff))
    rel_err = np.linalg.norm(diff) / np.linalg.norm(target)
    corr = np.corrcoef(pred.flatten(), target.flatten())[0, 1]
    return {
        'name': name,
        'mse': mse,
        'rmse': rmse,
        'mae': mae,
        'rel_err': rel_err,
        'correlation': corr
    }

# Metrics for pytorch_recons
print("\n--- From PYTORCH RECONSTRUCTION ---")
m1 = compute_metrics(mumott_fwd_from_pytorch, pytorch_fwd_from_pytorch, "mumott vs pytorch forward")
print(f"mumott vs pytorch forward:")
print(f"  Relative Error: {m1['rel_err']:.4%}")
print(f"  Correlation: {m1['correlation']:.6f}")
print(f"  MSE: {m1['mse']:.4e}")

m2 = compute_metrics(mumott_fwd_from_pytorch, target_data, "mumott forward vs target")
print(f"\nmumott forward vs target:")
print(f"  Relative Error: {m2['rel_err']:.4%}")
print(f"  Correlation: {m2['correlation']:.6f}")
print(f"  MSE: {m2['mse']:.4e}")

m3 = compute_metrics(pytorch_fwd_from_pytorch, target_data, "pytorch forward vs target")
print(f"\npytorch forward vs target:")
print(f"  Relative Error: {m3['rel_err']:.4%}")
print(f"  Correlation: {m3['correlation']:.6f}")
print(f"  MSE: {m3['mse']:.4e}")

# Metrics for mumott_recons
print("\n--- From MUMOTT RECONSTRUCTION ---")
m4 = compute_metrics(mumott_fwd_from_mumott, pytorch_fwd_from_mumott, "mumott vs pytorch forward")
print(f"mumott vs pytorch forward:")
print(f"  Relative Error: {m4['rel_err']:.4%}")
print(f"  Correlation: {m4['correlation']:.6f}")
print(f"  MSE: {m4['mse']:.4e}")

m5 = compute_metrics(mumott_fwd_from_mumott, target_data, "mumott forward vs target")
print(f"\nmumott forward vs target:")
print(f"  Relative Error: {m5['rel_err']:.4%}")
print(f"  Correlation: {m5['correlation']:.6f}")
print(f"  MSE: {m5['mse']:.4e}")

m6 = compute_metrics(pytorch_fwd_from_mumott, target_data, "pytorch forward vs target")
print(f"\npytorch forward vs target:")
print(f"  Relative Error: {m6['rel_err']:.4%}")
print(f"  Correlation: {m6['correlation']:.6f}")
print(f"  MSE: {m6['mse']:.4e}")
```

```python
# ============================================================================
# Summary Table
# ============================================================================
print("\n" + "=" * 80)
print("SUMMARY: Forward Model Consistency")
print("=" * 80)

print("\n{:<35} {:>15} {:>15}".format("", "pytorch_recons", "mumott_recons"))
print("-" * 65)
print("{:<35} {:>15.4%} {:>15.4%}".format(
    "mumott vs pytorch (rel err)", m1['rel_err'], m4['rel_err']))
print("{:<35} {:>15.6f} {:>15.6f}".format(
    "mumott vs pytorch (corr)", m1['correlation'], m4['correlation']))
print("-" * 65)
print("{:<35} {:>15.4%} {:>15.4%}".format(
    "mumott fwd vs target (rel err)", m2['rel_err'], m5['rel_err']))
print("{:<35} {:>15.4%} {:>15.4%}".format(
    "pytorch fwd vs target (rel err)", m3['rel_err'], m6['rel_err']))
print("-" * 65)
print("{:<35} {:>15.6f} {:>15.6f}".format(
    "mumott fwd vs target (corr)", m2['correlation'], m5['correlation']))
print("{:<35} {:>15.6f} {:>15.6f}".format(
    "pytorch fwd vs target (corr)", m3['correlation'], m6['correlation']))

print("\n" + "=" * 80)
print("KEY FINDINGS:")
print("=" * 80)

# Check if forward models agree better on mumott_recons than pytorch_recons
if m1['rel_err'] > m4['rel_err'] * 1.5:  # 1.5x threshold
    print(f"\n1. Forward models DISAGREE more on pytorch_recons ({m1['rel_err']:.1%})")
    print(f"   than on mumott_recons ({m4['rel_err']:.1%})")
    print(f"   → The pytorch reconstruction has different structure than mumott's")

# Check which reconstruction is more data-consistent
print(f"\n2. Data consistency comparison:")
print(f"   pytorch_recons: pytorch_fwd achieves {m3['rel_err']:.1%} error vs target")
print(f"   mumott_recons:  mumott_fwd achieves {m5['rel_err']:.1%} error vs target")

if m3['rel_err'] < m5['rel_err']:
    print(f"   → pytorch_recons is MORE data-consistent when using pytorch forward")
else:
    print(f"   → mumott_recons is MORE data-consistent when using mumott forward")

# Check the cross-application
print(f"\n3. Cross-application insight:")
print(f"   When applying pytorch_fwd to mumott_recons: {m6['rel_err']:.1%} error")
print(f"   When applying mumott_fwd to pytorch_recons: {m2['rel_err']:.1%} error")

print("\n" + "=" * 80)
```

## 8. Hybrid Approach: SAXSProjector + forward_quadrature

Now we test a hybrid approach combining:
- **mumott's SAXSProjector** for the spatial projection (fast ASTRA-based projection)
- **PyTorch's forward_quadrature** for the SH forward (GPU-accelerated spherical harmonics)

This should give us the best of both worlds: fast projection from mumott and fast SH from our CUDA-optimized implementation.

```python
# Import the saxs_pytorch wrapper
from smartt.optimization.saxs_pytorch import build_saxs_projector, SAXSProjectorLayer

# Build the SAXSProjector-based projector (wraps mumott's fast ASTRA projection)
# Note: use_cuda=False since SAXSProjectorCUDA is currently disabled in saxs_pytorch.py
saxs_projector_fn = build_saxs_projector(geometry, use_cuda=True, device=device)

print("SAXS Projector built successfully")
print(f"Device: {device}")
```

```python
# Define the hybrid forward function
def hybrid_forward(field: torch.Tensor, q_vectors: torch.Tensor, proj_matrix: torch.Tensor) -> torch.Tensor:
    """
    Hybrid forward: SAXSProjector for spatial projection + forward_quadrature for SH.
    
    Steps:
    1. Apply spatial projection using SAXSProjector (via saxs_pytorch wrapper)
    2. Apply SH forward using forward_quadrature
    
    Parameters
    ----------
    field : torch.Tensor
        SH coefficients, shape (X, Y, Z, C)
    q_vectors : torch.Tensor
        Probed coordinates from geometry
    proj_matrix : torch.Tensor
        SH projection matrix
    
    Returns
    -------
    torch.Tensor
        Projected images, shape (n_projections, H, W, n_directions)
    """
    # Step 1: Spatial projection using SAXSProjector
    # Input: (X, Y, Z, C) -> Output: (n_projections, H, W, C)
    projected_coeffs = saxs_projector_fn(field)
    
    # Step 2: SH forward using forward_quadrature
    # Input: (n_projections, H, W, C) -> Output: (n_projections, H, W, n_directions)
    projected_images = forward_quadrature(
        q_vectors,
        projected_coeffs,
        ell_max=ELL_MAX,
        projection_matrix=proj_matrix
    )
    
    return projected_images

print("Hybrid forward function defined")
```

## 9. PyTorch LBFGS Optimizer

Now let's try using PyTorch's LBFGS optimizer instead of AdamW. LBFGS is a quasi-Newton method that typically converges faster than first-order methods like Adam, similar to what mumott uses internally.

```python
def run_pytorch_lbfgs_reconstruction(dc, ell_max, maxiter, reg_weight=0.0, reg_type='none', 
                                      volume=None, verbose=True, history_size=10, line_search_fn='strong_wolfe'):
    """Run PyTorch LBFGS reconstruction and return results with timing.
    
    Parameters
    ----------
    dc : DataContainer
        The data container with projections
    ell_max : int
        Maximum spherical harmonic degree
    maxiter : int
        Maximum number of LBFGS iterations
    reg_weight : float
        Regularization weight
    reg_type : str
        Type of regularization ('none', 'laplacian')
    volume : torch.Tensor, optional
        Initial volume (if None, starts from zeros)
    verbose : bool
        Print progress
    history_size : int
        LBFGS history size (number of previous steps to use for Hessian approximation)
    line_search_fn : str
        Line search function ('strong_wolfe' or None)
    """
    from torch.optim import LBFGS as TorchLBFGS
    

    geometry = dc.geometry
    volume_shape = geometry.volume_shape
    num_coeffs = (ell_max // 2 + 1) * (ell_max + 1)
    
    # Get target data
    target_data = torch.tensor(
        np.stack([p.data for p in dc.projections]).astype(np.float32),
        device=device, dtype=torch.float32
    )
    
    # Create basis set and projection matrix
    basis_set = SphericalHarmonics(ell_max=ell_max, probed_coordinates=geometry.probed_coordinates)
    projection_matrix = torch.tensor(
        basis_set.projection_matrix.astype(np.float32),
        device=device, dtype=torch.float32
    )
    
    # Build projector
    projector = build_mumott_projector(geometry, device=device)
    
    # Initialize volume
    if volume is None:
        volume = torch.zeros(
            (*volume_shape, num_coeffs),
            device=device, dtype=torch.float32, requires_grad=True
        )
    else:
        volume = volume.to(device).clone().detach().requires_grad_(True)
    
    # LBFGS optimizer
    optimizer = TorchLBFGS(
        [volume], 
        lr=1.0,  # LBFGS typically uses lr=1.0
        max_iter=20,  # Inner iterations per step
        history_size=history_size,
        line_search_fn=line_search_fn,
        tolerance_grad=1e-7,
        tolerance_change=1e-9
    )
    
    losses = []
    iteration_count = [0]  # Use list to allow modification in closure
    
    def closure():
        """Closure function required by LBFGS."""
        optimizer.zero_grad()
        
        # Forward pass
        spatial_proj = projector(volume)
        predicted = forward_quadrature(
            geometry.probed_coordinates,
            spatial_proj,
            ell_max=ell_max,
            projection_matrix=projection_matrix
        )
        
        # Compute loss (sum of squared errors to match mumott)
        data_loss = torch.sum((predicted - target_data) ** 2)
        
        # Regularization
        reg_loss = torch.tensor(0.0, device=device)
        if reg_weight > 0 and reg_type == 'laplacian':
            laplacian = (
                torch.roll(volume, 1, 0) + torch.roll(volume, -1, 0) +
                torch.roll(volume, 1, 1) + torch.roll(volume, -1, 1) +
                torch.roll(volume, 1, 2) + torch.roll(volume, -1, 2) - 6 * volume
            )
            reg_loss = reg_weight * torch.sum(laplacian ** 2)
        
        loss = data_loss + reg_loss
        
        # Backward pass
        loss.backward()
        
        return loss
    
    if verbose:
        print(f"Running PyTorch LBFGS ({maxiter} iterations)...")
    
    torch.cuda.synchronize() if torch.cuda.is_available() else None
    start = time.time()
    
    for i in range(maxiter):
        # LBFGS step
        loss = optimizer.step(closure)
        losses.append(loss.item())
        
        if verbose and (i % 10 == 0 or i == maxiter - 1):
            print(f"Iteration {i:3d}: Loss = {loss.item():.6e}")
    
    torch.cuda.synchronize() if torch.cuda.is_available() else None
    total_time = time.time() - start
    
    return {
        'reconstruction': volume.detach().cpu(),
        'time': total_time,
        'final_loss': losses[-1],
        'losses': losses,
        'iterations': maxiter
    }

print("PyTorch LBFGS reconstruction function defined.")
```

## Summary: Why Reconstructions Differ

**The loss functions are NOT the same!**

Even though both optimizers minimize $\|A(x) - y\|^2$, the forward operators $A$ are different:
- `A_mumott`: Uses mumott's SAXSProjector + SphericalHarmonics.forward()
- `A_pytorch`: Uses ASTRA projector + forward_quadrature()

**Key differences:**
1. **Different interpolation**: ASTRA uses different interpolation than mumott's CPU projector
2. **Different adjoints**: The gradient $\nabla_x \|A(x) - y\|^2 = 2 A^*(A(x) - y)$ depends on the adjoint $A^*$
3. **Compounding errors**: Small forward errors (~10%) lead to large adjoint differences

**Result:**
- The pytorch optimizer found $x^*_{pytorch}$ where $\nabla_{pytorch} L = 0$
- But at $x^*_{pytorch}$, $\nabla_{mumott} L \neq 0$ (and vice versa)
- Both solutions minimize their respective loss, but they are geometrically different

**To make reconstructions match:**
1. Use the EXACT same forward model (e.g., both use pytorch projector)
2. Or, use the EXACT same projector implementation (match ASTRA settings with mumott)
