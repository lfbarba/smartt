```python
# Fig. 5(b), Eq. 7: cross-correlation of mean 

def compute_cc(recon, gt):
    mean_recon = np.mean(recon)
    mean_gt = np.mean(gt)
    numerator = np.sum((recon - mean_recon) * (gt - mean_gt))
    denominator = np.sqrt(np.sum((recon - mean_recon) ** 2)) * np.sqrt(np.sum((gt - mean_gt) ** 2))
    cc = numerator / denominator
    return cc
```
